```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
tips.head()

```

```python
flights.head()
```

```python
tc = tips.corr()
```

```python
tc
```

```python
sns.heatmap(tc,annot = True,fmt ='.2g')
```

```python
flights
```

```python
fp = flights.pivot_table(index = 'month',columns = 'year',values = 'passengers')

```

```python
fp
```

```python
sns.heatmap(fp) # original heatmap
```

```python
sns.heatmap(fp,annot = True,fmt = '0g',cmap = 'magma',linecolor = 'white',linewidths = 1) 
```

```python
sns.clustermap(fp) #original
```

```python
sns.clustermap(fp,cmap = 'magma',annot = True,fmt = '0g',linecolor = 'white',linewidth = 1)
```

```python
sns.clustermap(fp,cmap = 'magma',annot = True,fmt = '0.0g',linecolor = 'white',linewidth = 1,standard_scale = 1)# Std scallin-2
```

```python
%matplotlib inline
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```
